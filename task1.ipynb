```python
# This script fetches weather data for multiple cities using the OpenWeatherMap API.
# It retrieves the current temperature and humidity for each city and stores the results in a list.
import requests

API_KEY = "YOUR_API_KEY"
CITIES = [
    {"name": "London", "lat": 51.5074, "lon": -0.1278},
    {"name": "New York", "lat": 40.7128, "lon": -74.0060},
    {"name": "Tokyo", "lat": 35.6895, "lon": 139.6917},
    {"name": "Mumbai", "lat": 19.0760, "lon": 72.8777},
    {"name": "Paris", "lat": 48.8566, "lon": 2.3522},
    {"name": "Sydney", "lat": -33.8688, "lon": 151.2093},
    {"name": "Moscow", "lat": 55.7558, "lon": 37.6173},
    {"name": "Beijing", "lat": 39.9042, "lon": 116.4074},
    {"name": "Cairo", "lat": 30.0444, "lon": 31.2357},
    {"name": "Rio de Janeiro", "lat": -22.9068, "lon": -43.1729},
    {"name": "Los Angeles", "lat": 34.0522, "lon": -118.2437},
    {"name": "Toronto", "lat": 43.6532, "lon": -79.3832},
    {"name": "Berlin", "lat": 52.5200, "lon": 13.4050},
    {"name": "Dubai", "lat": 25.2048, "lon": 55.2708},
    {"name": "Singapore", "lat": 1.3521, "lon": 103.8198},
    {"name": "Johannesburg", "lat": -26.2041, "lon": 28.0473},
    {"name": "Mexico City", "lat": 19.4326, "lon": -99.1332},
    {"name": "Istanbul", "lat": 41.0082, "lon": 28.9784},
    {"name": "Seoul", "lat": 37.5665, "lon": 126.9780},
    {"name": "Bangkok", "lat": 13.7563, "lon": 100.5018},
    {"name": "Buenos Aires", "lat": -34.6037, "lon": -58.3816},
    {"name": "Rome", "lat": 41.9028, "lon": 12.4964},
    {"name": "Madrid", "lat": 40.4168, "lon": -3.7038},
    {"name": "Jakarta", "lat": -6.2088, "lon": 106.8456},
    {"name": "Lagos", "lat": 6.5244, "lon": 3.3792},
    {"name": "Chicago", "lat": 41.8781, "lon": -87.6298},
    {"name": "Hong Kong", "lat": 22.3193, "lon": 114.1694},
    {"name": "San Francisco", "lat": 37.7749, "lon": -122.4194},
    {"name": "Kuala Lumpur", "lat": 3.1390, "lon": 101.6869},
    {"name": "Nairobi", "lat": -1.2921, "lon": 36.8219}
]
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"

weather_data = []

for city in CITIES:
    params = {
        "lat": city["lat"],
        "lon": city["lon"],
        "appid": API_KEY,
        "units": "metric"
    }
    response = requests.get(BASE_URL, params=params)
    data = response.json()
    
    if response.status_code == 200:
        temp = data["main"]["temp"]
        humidity = data["main"]["humidity"]
        weather_data.append({
            "city": city["name"],
            "temperature": temp,
            "humidity": humidity
        })
    else:
        print(f"Failed to fetch data for {city['name']}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#data for ploting
cities = [entry["city"] for entry in weather_data]
temps = [entry["temperature"] for entry in weather_data]
humidities = [entry["humidity"] for entry in weather_data]

sns.set(style="whitegrid")

#temperature bar chart
plt.figure(figsize=(12, max(8, len(cities) * 0.4)))
sns.barplot(y=cities, x=temps, palette="coolwarm")
plt.title("Temperature Comparison Across Cities (C)")
plt.xlabel("Temperature (C)")
plt.ylabel("City")
plt.tight_layout()
plt.show()

#humidity bar chart
plt.figure(figsize=(12, max(8, len(cities) * 0.4)))
sns.barplot(y=cities, x=humidities, palette="Blues_d")
plt.title("Humidity Comparison Across Cities (%)")
plt.xlabel("Humidity (%)")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
#temperature vs humidity scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=temps, y=humidities, hue=cities, s=100)
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# temperature distribution pie chart
plt.figure(figsize=(13, 13))
plt.pie(temps, labels=cities, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("coolwarm", len(cities)))
plt.title("Temperature Distribution by City")
plt.tight_layout()
plt.show()
```

```python
#temperature trend line plot
sorted_data = sorted(weather_data, key=lambda x: x['temperature'])
sorted_cities = [item['city'] for item in sorted_data]
sorted_temps = [item['temperature'] for item in sorted_data]

plt.figure(figsize=(12, max(8, len(cities) * 0.4)))
sns.lineplot(x=sorted_temps, y=sorted_cities, marker='o', color='orange')
plt.title("Sorted Temperature Trend")
plt.ylabel("City")
plt.xlabel("Temperature (C)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame(weather_data)
df.set_index('city', inplace=True)

#heatmap of weather metrics
plt.figure(figsize=(12, max(8, len(cities) * 0.4)))
sns.heatmap(df, annot=True, cmap="YlGnBu", fmt=".1f")
plt.title("Heatmap of Weather Metrics")
plt.tight_layout()
plt.show()
```
